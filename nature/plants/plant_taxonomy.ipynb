# Match genera names to a higher level in plant taxonomy

This is because for the work I'm doing for a viz I need to coarse data - I've got too many genera

```python
df_taxons = pd.read_csv('backbone-current/Taxon.tsv', sep='\t')    # from the taxonomy sheet
```

```python
# take only plants - I think there won't be shared names between kindgoms but good to reduce upfront
df_taxons = df_taxons[df_taxons.kingdom=='Plantae']
```

```python
df_taxons = df_taxons.dropna(subset=['genus'])   # drop all those with no genus (there are) as I'll search by it
```

```python
# lower case to all cols of interest - easier to deal with, not worrying on case
df_taxons['genus']  = df_taxons['genus'].str.lower()
df_taxons['family']  = df_taxons['family'].str.lower()
df_taxons['order']  = df_taxons['order'].str.lower()
df_taxons['class']  = df_taxons['class'].str.lower()
df_taxons['phylum']  = df_taxons['phylum'].str.lower()
```

```python
# lower case on the genera sheer as well

df_genera['Genus'] = df_genera['Genus'].str.lower()
```

```python
# get info on all the genera I have, build a new sheet with all taxonomy info

genera = list(df_genera['Genus'])
len(genera)
```

```python
# genus acrocarpus has 2 families! See how big in number of lines
df_taxons[df_taxons.genus=='alkanna'].groupby('family').count() \
     .sort_values('genus', ascending=False).reset_index().iloc[0]['family']

#df_taxons[(df_taxons.genus=='acrocarpus') & (df_taxons.family=='gelidiaceae')]


df_taxons[df_taxons.genus=='alkanna'].groupby(['family', 'order', 'class', 'phylum']).count() \
     .sort_values('genus', ascending=False).reset_index()[['family', 'order', 'class', 'phylum', 'genus']]
```

```python
# # THIS IS just to check
# # there are some genera which map to more than one family, or more than one order, or class etc
# # it'll likely be mistakes - we take the family/order/class/... where most of items for that genus map to (majority)
# (in next cell)

print('genus - family')
for genus in genera:
    if len(df_taxons[df_taxons.genus==genus].groupby('family').count()) > 1:
        print(genus)
        
```

```python
# we're using the most populated of the higher levels for that genus as per above

phylum, class_, order, family, genus = [], [], [], [], []
i = 0
for g in genera[:]:    # search in taxon sheet
    
    # check if there first - otehrwise do not cosider at all
    if len(df_taxons[df_taxons.genus==g]) > 0:
        
        genus.append(g)
        
        max_ = df_taxons[df_taxons.genus==g].groupby(['family', 'order', 'class', 'phylum']) \
                 .count() \
                 .sort_values('genus', ascending=False).reset_index().iloc[0]
        
        family.append(max_['family'])
        order.append(max_['order'])
        class_.append(max_['class'])
        phylum.append(max_['phylum'])
    
    i += 1
    
    if i % 50 == 0:
        print(i)
```

```python
df = pd.DataFrame({
    'genus': genus,
    'family': family,
    'order': order,
    'class': class_,
    'phylum': phylum
})
```
