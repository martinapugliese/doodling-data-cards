```python
import pandas as pd

from matplotlib import pyplot as plt
import matplotlib

%matplotlib inline
```

```python
df = pd.read_csv('matches.csv')
```

```python
# add col for tot cards/ tot matches played

df['cards_to_matches'] = df['Tot cards'] / df['Matches played']
```

```python
df
```

```python
df[df['Matches played'] >= 30].plot('Matches played', 'cards_to_matches', kind='scatter')

plt.grid()
plt.show();
```
