# Cocktails analysis

* combos of 2 alcoholic ingredients for each cocktail

```python
import json
import itertools
import pandas as pd
import collections
```

```python
# read data
d_unforgettables = json.load(open('unforgettables.json'))  
d_contemporary = json.load(open('contemporary_classics.json')) 
d_new  = json.load(open('new_era.json')) 

d = d_unforgettables.copy()
d.update(d_contemporary)
d.update(d_new)

booze = pd.read_csv('boozes.csv')
```

```python
len(d), len(d_unforgettables), len(d_contemporary), len(d_new), len(booze)
```

```python
booze.head()
```

```python
# remove the non alcoholic ingredients from lists of ingredients
# get the polished booze name

d_boozy = {}
for k in d:
    d_boozy[k] = []
    for ingredient in d[k]:
        if len(booze[booze['Booze']==ingredient]) >0:
            d_boozy[k].append(booze[booze['Booze']==ingredient].iloc[0]['Polished_booze'])
```

```python
d_boozy['bacardi']
```

```python
# for each cocktail compute all combos of 2 ingredients, using polshed version
# note - if there is only one alcoholic ingredient it won't appear (must be >= 2)

couples = []
for cocktail_name in d_boozy:
    print(cocktail_name)
    combos = itertools.combinations(d_boozy[cocktail_name], 2)
    for combo in combos:
        couples.append(tuple(sorted(combo)))   # put in alphabetic order to avoid repetition
        print(combo)
    print('\n')
    
# note could put a set but wpouldn't be hashable for counting purposes
```

```python
couples
```

```python
c = collections.Counter(couples)

c.most_common()
```

```python
booze['Polished_booze'].nunique()
```

```python
unique_boozes = []
for t_ in c.most_common():
    for item in t_[0]:
        unique_boozes.append(item)
```
