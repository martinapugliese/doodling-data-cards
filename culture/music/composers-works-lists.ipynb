## Beethoven

```python
df = pd.read_csv('Composers\' works - Beethoven.csv')

len(df)
```

```python
## manipulate

# remove missing date ones 
df = df[(~df['start_year'].isna()) & (~df['start_year'].isin(['—', '-']))]

# transform the start_year in an int
df['start_year'] = df['start_year'].astype(int)

# add col for lustrum (lustro)
df['lustrum'] = df['start_year'].apply(lambda row: divmod(row, 10)[0] * 10 + 5 * divmod(divmod(row, 10)[1], 5)[0])

df = df[~df.key_mod.isna()]
```

```python
len(df)

df.head(2)

df.dtypes
```

```python
df.groupby('key_mod').count()

set(df['key_mod'])
```
