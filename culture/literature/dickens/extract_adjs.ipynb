```python
from textblob import TextBlob
from collections import Counter
from nltk.corpus import stopwords
```

```python
f = open('pictures_from_Italy.txt')
```

```python
blob = TextBlob(f.read())
```

```python
blob.sentiment
```

```python
len(blob.words), len(set(blob.words))
```

```python
blob.word_counts
```

```python
adjs = []
for item  in blob.tags:
    if item[1]=='JJ':
        adjs.append(item[0].lower())
```

```python
len(adjs), len(set(adjs))
```

```python
c = Counter(adjs)
```

```python
c.most_common(100)
```
